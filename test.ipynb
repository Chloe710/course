```python
# 函数定义（内部代码不执行）
def check_input():
    while True:
        num = input('请输入一个数值：')
   
        if num.isdigit() == True:
            break

    v=int(num)
    return v
#输入 n个数字,找到其中最大和最小值
n=5
#列表保存用户输入
numbers = []
for i in range(n):
#循环中执行函数，接收结果
    result=check_input()
    numbers.append(result)
print(f"最大值：{max(numbers)} 最小值：{min(numbers)}")
```

```python
name={'a':111,'b':222,'c':333,'d':444}
a=int(input('please:'))
while True:
    if a ==1:
        apn=input('please input the name you want to add:\t')
        apt=input('please input the tel you want to add:\t')
        name[apn]=apt
        
    elif a ==2:
        for key in name:
            print(f'{key} : {name[key]}')
    elif a==3:
        del_key=None
        del_inp=input('please input the name you want to delete: ')
        for key in name:
            if key==del_inp:
                del_key=key
            
        if del_key is not None:
            del name[del_key]
            print(name)
        else:
            print(f'cannot find {del_inp}')
            
    elif a==4:
        fix_key=None
        fix_n=input('please input the name you want to fix: ')
        fix_t=input('please input the tel you want to fix: ')
        for key in name:
            if key==fix_n:
                fix_key=key
            
        if fix_key is not None:
            name[fix_key]=fix_t
            print(name)
        else:
            print(f'cannot find {fix_n}')      

    elif a==0:
        print('The program is done!')
        break
    else:
        print('please input the correct number!')
       
```

```python
import time
from random import randint as rd
ren=rd(100,150)
ji=rd(100,150)
reng=rd(30,50)
jig=rd(30,50)
times=[1,2,3]
again=True
while again:
    for i in range(3):
        print(f'--------------round {times[i]}-------------------')
        print(f'您的血量为{ren},对手的血量为{ji}')
        print('---------------------------------')
        time.sleep(2)
        ren=ren-jig
        ji=ji-reng
        print(f'您发起了{reng}攻击,对手血量为{ji}')
        print(f'对手发起了{jig}攻击,您的血量为{ren}')
        print('---------------------------------')
        time.sleep(2)
    #if ren <0 or ji<0:
       # break
    if ren > ji:
        print(f'您win!')
    elif ren == ji:
        print(f'您和对手打平!')
    else:
        print(f'that`s too bad!对手win!')
    a1 = input('要继续游戏吗,请输入n退出,输入其他继续：')
    if a1=='n':
        again=False
    else:
         again=True   
```

```python
import time
from random import randint as rd

times=[1,2,3,4,5]
reds=[]
greens=[]
rsum=0
gsum=0
count=0
while True:
    for i in range(5):
        red=rd(1,3)
        green=rd(1,3)
        print(f'--------------round {times[i]}-------------------')
        print(f'红队进球，得{red}分')
        reds.append(red)
        rsum=sum(reds)
        print(f'红队总分为{rsum}')
        print(f'绿队进球，得{green}分')
        greens.append(green)
        gsum=sum(greens)
        print(f'绿队总分为{gsum}')
        time.sleep(2)
        count+=1
        if count==10:
            break
        
    if rsum > gsum:
        print(f' 红队win!')
    elif rsum == gsum:
        print(f'红队和绿队打平!')
    else:
        print(f'绿队win!')
    a1 = input('要继续游戏吗,请输入n退出,输入其他继续：')
    if a1=='n':
        break
    
```
